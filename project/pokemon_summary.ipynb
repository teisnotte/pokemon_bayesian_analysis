```python
import polars as pl
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pokemon_df = pl.read_csv("./data/Pokemon.csv")
pokemon_df.head()
```

```python
pokemon_df.sort(pl.col("Total"), descending=False).head(10)
```

```python
typing = ["Type 1", "Type 2"]
ability = ["HP", "Attack", "Defense", "Sp.Atk", "Sp.Def", "Speed"]
```

# 1. General Statistics for the Dataset

```python
print(pokemon_df[ability].describe())
```

```python
# Set the theme for the plot
sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

long_df = pokemon_df.unpivot(
    index=['Name'],
    on=ability,
    variable_name='Stat',
    value_name='Value'
)

# Initialize the FacetGrid object
pal = sns.cubehelix_palette(len(ability), rot=-.25, light=.7)
g = sns.FacetGrid(long_df, row="Stat", hue="Stat", aspect=9, height=0.75, palette=pal)

# Draw the densities for each stat
g.map(sns.kdeplot, "Value",
      bw_adjust=.5, clip_on=False,
      fill=True, alpha=1, linewidth=1.5)
g.map(sns.kdeplot, "Value", clip_on=False, color="w", lw=2, bw_adjust=.5)

# Add a horizontal line to each plot
g.map(plt.axhline, y=0, lw=2, clip_on=False)

# Define and use a function to label each plot
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, fontweight="bold", color=color,
            ha="left", va="center", transform=ax.transAxes)

g.map(label, "Value")

# Set the subplots to overlap for the ridge effect
g.figure.subplots_adjust(hspace=-.5)

# Remove axes details that are not needed
g.set_titles("")
g.set(yticks=[], ylabel="")
g.despine(bottom=True, left=True)
g.fig.suptitle('Overall Distribution of Pokémon Stats', y=1.02, fontsize=16)
g.savefig("./result/pokemon_stats_distribution.png", bbox_inches='tight', dpi=300)
```

```python
unique_typing = set(pokemon_df["Type 1"].unique().to_list() + pokemon_df["Type 2"].unique().to_list()) - set([None])

pokemon_df = pokemon_df.with_columns(
    pl.concat_list([pl.col("Type 1"), pl.col("Type 2")])
    .list.drop_nulls()
    .list.sort()
    .list.join(", ")
    .alias("CombinedTyping")
)
pokemon_df.head()

combined_typing = pokemon_df["CombinedTyping"].unique().to_list()
print(combined_typing)
```

```python
for typ in unique_typing:
    sub_type = pokemon_df.filter(
        (pl.col("Type 1") == typ) | (pl.col("Type 2") == typ)
    )
    
    print(f"Type {typ}: {sub_type.shape[0]} Pokémon")
    print(sub_type[ability].describe())

    long_df = sub_type.unpivot(
        index=['Name'],
        on=ability,
        variable_name='Stat',
        value_name='Value'
    )


    # Initialize the FacetGrid object
    pal = sns.cubehelix_palette(len(ability), rot=-.25, light=.7)
    g = sns.FacetGrid(long_df, row="Stat", hue="Stat", aspect=9, height=0.75, palette=pal)

    # Draw the densities for each stat
    g.map(sns.kdeplot, "Value",
        bw_adjust=.5, clip_on=False,
        fill=True, alpha=1, linewidth=1.5)
    g.map(sns.kdeplot, "Value", clip_on=False, color="w", lw=2, bw_adjust=.5)

    # Add a horizontal line to each plot
    g.map(plt.axhline, y=0, lw=2, clip_on=False)

    # Define and use a function to label each plot
    def label(x, color, label):
        ax = plt.gca()
        ax.text(0, .2, label, fontweight="bold", color=color,
                ha="left", va="center", transform=ax.transAxes)

    g.map(label, "Value")

    # Set the subplots to overlap for the ridge effect
    g.figure.subplots_adjust(hspace=-.5)

    # Remove axes details that are not needed
    g.set_titles("")
    g.set(yticks=[], ylabel="")
    g.despine(bottom=True, left=True)
    g.fig.suptitle(f'Overall Distribution of Pokémon Type {typ}', y=1.02, fontsize=16)
    g.savefig(f"./result/pokemon_stats_distribution_type_{typ}.png", bbox_inches='tight', dpi=300)
```

```python
def clr_transform(row: dict) -> list[float]:
    values = list(row.values())
    log_stats = np.log(values)
    geometric_mean = np.mean(log_stats)
    return list(log_stats - geometric_mean)

old_col = [
    'HP', 'Attack', 'Defense',
    'Sp.Atk', 'Sp.Def', 'Speed'
]

new_col = [
        'clr_HP', 'clr_Attack', 'clr_Defense',
        'clr_SpAtk', 'clr_SpDef', 'clr_Speed'
    ]

closed_pokemon_df = pokemon_df.with_columns(
    pl.col(old_col) / pl.col("Total")
)

clr_data = closed_pokemon_df.with_columns(
    pl.struct(old_col)
    .map_elements(clr_transform)
    .alias('clr_stats')
)

clr_data = clr_data.with_columns(
    pl.col('clr_stats').list.to_struct(fields=new_col)
).unnest('clr_stats')

clr_data.head()
```

```python
import plotly.express as px
from sklearn.decomposition import PCA

pca = PCA(n_components=5)
components = pca.fit_transform(clr_data[new_col].to_numpy())

labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}
fig = px.scatter_matrix(
    components,
    labels=labels,
    dimensions=range(4),
    color=clr_data['Type 1'],
    width=800,
    height=800,
    title="PCA of Pokémon Stats (CLR Transformed)",
)
fig.update_traces(diagonal_visible=False)
fig
fig.show()


exp_var_cumul = np.cumsum(pca.explained_variance_ratio_)

px.area(
    x=range(1, exp_var_cumul.shape[0] + 1),
    y=exp_var_cumul,
    labels={"x": "# Components", "y": "Explained Variance"},
    width=800,
    height=400,
)
```

```python
loadings = pca.components_
loadings_df = pl.DataFrame(loadings, schema=new_col)
print(loadings_df)
```
